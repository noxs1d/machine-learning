```python
cat_features = data.select_dtypes(include=['object']).columns
cat_features
```

```python
cat_features = ['gender', 'marital_status', 'job_position', 'education']
```

```python
plt.style.use('ggplot')

for column in cat_features:
    plt.figure(figsize=(8, 6))
    plt.subplot(121)
    data[column].value_counts().plot(kind='bar')
    plt.xlabel(column)
    plt.ylabel('number of customers')
    plt.title(column)
```

```python
np.number
```

```python
numeric_features = data.select_dtypes(include=np.number).columns
numeric_features
```

```python
for col in numeric_features:
    plt.figure(figsize=(8, 4))
    plt.subplot(121)
    sns.distplot(data[col], kde=True)
    plt.title(col)
```

```python
data['open_account_flg']
```

```python
for feature in cat_features:
    plt.figure(figsize=(12, 4))
    plt.subplot(121)
    sns.countplot(x=data[feature], hue=data['open_account_flg'])
    plt.title(feature)
    plt.xticks(rotation=90)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
for feature in cat_features:
    data[feature] = le.fit_transform(data[feature])
```

```python
data_all_cols = data.columns
data_all_cols = list(data_all_cols)
print(data_all_cols)
print(type(data_all_cols))
data_all_cols = data_all_cols.remove('living_region')
print(data_all_cols)
```
