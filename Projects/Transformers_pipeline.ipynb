```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I've been waiting for a HuggingFace course my whole life.")
```

```python
classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)
```

```python
generator = pipeline("text-generation")
generator("In this course, we will teach you how to")
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

model_name = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoTokenizer.from_pretrained(model_name)
```

```python
raw_text = ["I love programming, and interested in AI", "I hate read documetation of libs"]
out = model(raw_text, padding=True, truncation=True, return_tensors="pt")
print(out)
```

```python
from transformers import AutoModel

automodel = AutoModel.from_pretrained(model_name)
output = automodel(**out)
print(output.last_hidden_state.shape)
```

```python
ClassifierModel = AutoModelForSequenceClassification.from_pretrained(model_name)
out = ClassifierModel(**out)
print(out.logits)
```

```python
import torch

predictions = torch.nn.functional.softmax(out.logits, dim=-1)
print(predictions)
```
