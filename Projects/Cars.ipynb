```python
from sklearn.preprocessing import OneHotEncoder
one_hot_car= OneHotEncoder(sparse_output=False)
```

```python
car_encoded=one_hot_car.fit_transform(car_data_objects)
```

```python
car_data_num=car_data.select_dtypes(exclude="object")
```

```python
car_data_num
```

```python
categorical_columns=car_data_objects.columns.to_list()
```

```python
df_cars=pd.DataFrame(car_encoded, columns=one_hot_car.get_feature_names_out(categorical_columns))
```

```python
car_datas=pd.concat([car_data_num,df_cars], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(car_datas.drop(["Selling_price"], axis=1), car_datas["Selling_price"], test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()

lin_reg.fit(X_train, y_train)
```

```python
predictions=lin_reg.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import mean_squared_error
lin_mse=mean_squared_error(y_test, predictions)
```

```python
lin_rmse=np.sqrt(lin_mse)
lin_rmse
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(X_train, y_train)
```

```python
tree_predictions=tree_reg.predict(X_test)
tree_rmse=np.sqrt(mean_squared_error(tree_predictions, y_test))
tree_rmse
```

```python
from sklearn.model_selection import cross_val_score
lin_scores = cross_val_score(
    lin_reg,
    X_train,
    y_train,
    scoring="neg_mean_squared_error",
    cv=5
)
```

```python
np.sqrt(-lin_scores)
```
